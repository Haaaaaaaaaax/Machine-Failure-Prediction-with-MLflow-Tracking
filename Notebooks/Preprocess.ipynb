```python
from sklearn.preprocessing import StandardScaler
```

**We will only scale numeric numbers**

```python
numeric_cols = ['footfall', 'AQ', 'USS', 'CS', 'VOC', 'RP', 'IP', 'Temperature']
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[numeric_cols])

```

```python
df_scaled = pd.DataFrame(df_scaled, columns=numeric_cols)
```

```python
df_scaled.head(10)
```

```python
df_scaled['tempMode'] = df['tempMode']
df_scaled['fail'] = df['fail']
```

```python
df_scaled.to_csv('Data/data_final_scaled.csv', index=False)
```
